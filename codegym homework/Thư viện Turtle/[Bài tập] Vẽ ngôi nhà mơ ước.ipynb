```python
# Create a scene with a house, a tree, and a sun
# Based on the picture at http://dragonometry.net/blog/?p=566

import turtle
import math

# Set the background color
screen = turtle.Screen()
screen.bgcolor("skyblue")

# Create our turtle
t = turtle.Turtle()
t.color("black")
t.speed(10)

# Draw and fill the front of the house
t.penup() 
t.goto(-150, -120)
t.pendown()
t.fillcolor("blue")
t.begin_fill()

t.forward(100)
t.left(90)
t.forward(110)
t.left(90)
t.end_fill()

# Draw and fill the front door
t.penup()
t.goto(-120, -120)
t.pendown()
t.fillcolor("lightgreen")
t.begin_fill()

t.forward(40)
t.left(90)
t.forward(60)
t.left(90)
t.forward(40)
t.left(90)
t.forward(60)
t.left(90)
t.end_fill()
    
# Front roof
t.penup()
t.goto(-150, -10)
t.pendown()
t.fillcolor("magenta")
t.begin_fill()
t.forward(100)
t.left(135)
t.forward(100 / math.sqrt(2))
t.left(90)
t.forward(100 / math.sqrt(2))
t.left(135)
t.end_fill()

# Side of the house
t.penup()
t.goto(-50, -120)
t.pendown()
t.fillcolor("yellow")
t.begin_fill()
t.left(30)
t.forward(60)
t.left(60)
t.forward(110)
t.left(120)
t.forward(60)
t.left(60)
t.forward(110)
t.left(90)
t.end_fill()


# Window
t.penup()
t.goto(-30, -60)
t.pendown()
t.fillcolor("brown")
t.begin_fill()
t.left(30)
t.forward(20)
t.left(60)
t.forward(30)
t.left(120)
t.forward(20)
t.left(60)
t.forward(30)
t.left(90)
t.end_fill()

# Side roof
t.penup()
t.goto(-50, -10)
t.pendown()
t.fillcolor("orange")
t.begin_fill()
t.left(30)
t.forward(60)
t.left(105)
t.forward(100 / math.sqrt(2))
t.left(75)
t.forward(60)
t.left(105)
t.forward(100 / math.sqrt(2))
t.left(45)
t.end_fill()

# Tree base
t.penup() 
t.goto(100, -130)
t.pendown()
t.fillcolor("brown")
t.begin_fill()

t.forward(20)
t.left(90)
t.forward(40)
t.left(90)
t.forward(20)
t.left(90)
t.forward(40)
t.left(90)
t.end_fill()

# Tree top
t.penup() 
t.goto(65, -90)
t.pendown()
t.fillcolor("lightgreen")
t.begin_fill()
t.forward(90)
t.left(135)
t.forward(90 / math.sqrt(2))
t.left(90)
t.forward(90 / math.sqrt(2))
t.left(135)
t.end_fill()

t.penup() 
t.goto(70, -45)
t.pendown()
t.fillcolor("lightgreen")
t.begin_fill()
t.forward(80)
t.left(135)
t.forward(80 / math.sqrt(2))
t.left(90)
t.forward(80 / math.sqrt(2))
t.left(135)
t.end_fill()

t.penup() 
t.goto(75, -5)
t.pendown()
t.fillcolor("lightgreen")
t.begin_fill()
t.forward(70)
t.left(135)
t.forward(70 / math.sqrt(2))
t.left(90)
t.forward(70 / math.sqrt(2))
t.left(135)
t.end_fill()

# Sun
t.penup() 
t.goto(55, 110)
t.fillcolor("yellow")
t.pendown()
t.begin_fill()
t.circle(24)
t.end_fill()

# Sun rays
t.penup()
t.goto(55, 134)
t.pendown()

t.penup()
t.forward(24)
t.pendown()
t.forward(25)
t.penup()
t.backward(25 + 24)
t.left(90)
t.penup()
t.forward(24)
t.pendown()
t.forward(25)
t.penup()
t.backward(25 + 24)
t.left(90)
t.penup()
t.forward(24)
t.pendown()
t.forward(25)
t.penup()
t.backward(25 + 24)
t.left(90)
t.penup()
t.forward(24)
t.pendown()
t.forward(25)
t.penup()
t.backward(25 + 24)
t.left(90)

t.right(45)

t.penup()
t.forward(24)
t.pendown()
t.forward(15)
t.penup()
t.backward(15 + 24)
t.left(90)
t.penup()
t.forward(24)
t.pendown()
t.forward(15)
t.penup()
t.backward(15 + 24)
t.left(90)t.penup()
t.forward(24)
t.pendown()
t.forward(15)
t.penup()
t.backward(15 + 24)
t.left(90)
    
t.left(45)

# Put down some labels
t.color("black")
t.penup()
t.goto(-150, 70)
t.pendown()
t.write("House", None, None, "14pt bold")
t.penup()
t.goto(150, -150)
t.pendown()
t.write("Tree", None, None, "14pt bold")
t.penup()
t.goto(130, 150)
t.pendown()
t.write("Sun", None, None, "14pt bold")

t.done()
```

```python
class Patient:
    def __init__(self, name, age, height, weight):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight
        self.bmi = weight / height ** 2
        
    def calc_bmi(self):
        self.bmi = self.weight / self.height ** 2
```

```python
# BMI Calculator

from tkinter import *
from tkinter.messagebox import showinfo

window = Tk()
window.title("Healthcare System")
window.geometry("500x300")

# Label
intro_lbl = Label(window, text="Healthcare System")
my_weight_lbl = Label(window, text="My weight(kg)")
my_height_lbl = Label(window, text="My heihgt(m)")
my_name_lbl = Label(window, text="My name")
my_age_lbl = Label(window, text="My age")
search_lbl = Label(window, text="Search:")

# Entry
weight = Entry(window)
height = Entry(window)
name = Entry(window)
age = Entry(window)
search_name = Entry(window)

# list_patient được sử dụng để lưu trữ thông tin các bệnh nhân
list_patient = []

def add_patient():
    # lấy thông tin từ các entry
    weight_value = float(weight.get())
    height_value = float(height.get())
    name_value = name.get()
    age_value = int(age.get())
    
    # khởi tạo đối tượng
    patient = Patient(name_value, age_value, height_value, weight_value)
    
    # thêm đối tượng vào danh sách
    list_patient.append(patient)
    
    showinfo("Healthcare System", "Patient added successfully!")
    

def search_patient():
    search_name_value = search_name.get()
    for patient in list_patient:
        if search_name_value in patient.name:
            showinfo("Search result", f"{patient.name} - BMI: {patient.bmi} - Age: {patient.age}")
            break

def update_patient():            
    # lấy thông tin từ các entry
    weight_value = float(weight.get())
    height_value = float(height.get())
    name_value = name.get()
    age_value = int(age.get())
    
    for patient in list_patient:
        if name_value == patient.name:
            patient.age = age_value
            patient.height = height_value
            patient.weight = weight_value
            patient.calc_bmi()
            showinfo("Update result", f"Update information for patient {patient.name} successfully!")
            
def delete_patient():
    search_name_value = search.name.get()
    for i in range(len(list_patient)):
        if search_name_value == list_patient[i].name:
            del list_patient[i]
            showinfo("Action result", "Delete patient successfully")
            break
            
# Button
add_btn = Button(window, text="Add", command=add_patient)
search_btn = Button(window, text="Search", command=search_patient)
update_btn = Button(window, text="Update", command=update_patient)
delete_btn = Button(window, text="Delete", command=delete_patient)


intro_lbl.grid(row=0, column=0, columnspan=2)
my_name_lbl.grid(row=1,column=0),
name.grid(row=1,column=1)
my_age_lbl.grid(row = 2, column = 0)
age.grid(row = 2, column = 1)
my_weight_lbl.grid(row = 3, column = 0)
weight.grid(row=3, column=1)
my_height_lbl.grid(row=4, column=0)
height.grid(row=4, column=1)
add_btn.grid(row=5, column=1)
update_btn.grid(row=5, column=2)
delete_btn.grid(row=7, column=2)


search_lbl.grid(row=6, column=0)
search_name.grid(row=6, column=1)
search_btn.grid(row=6, column=2)

window.mainloop()
```

```python
from tkinter import *
from tkcalendar import Calendar
from datetime import datetime

# Create Object
root = Tk()

# Set geometry
root.geometry("400x400")

# Add Calendar
cal = Calendar(root, selectmode = 'day',
			year = 2020, month = 5,
			day = 22)

cal.pack(pady = 20)

checkin_date = None
checkout_date = None

def check_in():
	global checkin_date
	checkin_date = cal.get_date()
	checkin_label.config(text = "Check in date: " + checkin_date)

def check_out():
	global checkout_date
	checkout_date = cal.get_date()
	checkout_label.config(text = "Check out date: " + checkout_date)

def calculate_days():
    if checkin_date and checkout_date:
        date_format = "%m/%d/%y"
        checkin_datetime = datetime.strptime(checkin_date, date_format)
        checkout_datetime = datetime.strptime(checkout_date, date_format)
        days = (checkout_datetime - checkin_datetime).days
        total = days * 1000000
        days_label.config(text="Total: " + str(total) + " for" + str(days) + " day(s)" )

def update_customers():
	customers_listbox.delete(0, END)
	customers = ["Customer 1", "Customer 2", "Customer 3"]
	for customer in customers:
		customers_listbox.insert(END, customer)

# Add Button and Label
Button(root, text="Check in", command=check_in).pack(pady=10)
Button(root, text="Check out", command=check_out).pack(pady=10)
Button(root, text="Calculate days", command=calculate_days).pack(pady=10)
Button(root, text="Update customers", command=update_customers).pack(pady=10)

checkin_label = Label(root, text="Check in")
checkin_label.pack(pady=20)

checkout_label = Label(root, text="Check out")
checkout_label.pack(pady=20)

days_label = Label(root, text="")
days_label.pack(pady=20)

customers_listbox = Listbox(root)
customers_listbox.pack(pady=20)

# Execute Tkinter
root.mainloop()
```

```python
class Customer:
    def __init__(self, name, email, number, check_in, check_out, total):
        self.name = name
        self.email = email
        self.number = number
        self.check_in = check_in
        self.check_out = check_out
        self.total = total
```

```python
from tkinter import *
from tkcalendar import Calendar
from datetime import datetime
from tkinter.messagebox import showinfo

window = Tk()
window.title('STORE')
window.geometry('800x500')

# Label
intro_lbl = Label(window, text='Storge')

my_name_lbl = Label(window, text='Name')
email_lbl = Label(window, text='Email')
number_lbl = Label(window, text ='Phone number') 
search_name_lbl = Label(window, text = 'Search customer')

#Entry
name = Entry(window)
email = Entry(window)
number = Entry(window)


search_name = Entry(window)


#List_patient được sử dụng để lưu trữ thông tin các bệnh nhân
list_customer = []

#Lấy thông tin checkin, check out
# Add Calendar
cal = Calendar(window, selectmode = 'day',
			year = 2020, month = 5,
			day = 22)

cal.grid(row=8, column=1)

def check_in():
	global checkin_date
	checkin_date = cal.get_date()
	checkin_label.config(text = "Check in date: " + checkin_date)

def check_out():
	global checkout_date
	checkout_date = cal.get_date()
	checkout_label.config(text = "Check out date: " + checkout_date)

def calc_total():
    global total
    if checkin_date and checkout_date:
        date_format = "%m/%d/%y"
        checkin_datetime = datetime.strptime(checkin_date, date_format)
        checkout_datetime = datetime.strptime(checkout_date, date_format)
        days = (checkout_datetime - checkin_datetime).days
        total = days * 100000
        total_label.config(text="Total bill is " + str(total) + " VND, for " + str(days) + " day(s)")

def add_customer():
    #Lấy thông tin từ các entry:
    check_in_value = checkin_date
    check_out_value = checkout_date
    total_value = total
    name_value = name.get()
    email_value = email.get()
    number_value = number.get()
    
    #Khởi tạo đối tượng:
    customer = Customer(name_value, email_value, number_value, check_in_value, check_out_value, total_value)
    
    #Thêm đối tượng vào danh sách:
    list_customer.append(customer)
    
    showinfo("Add Customer", "Customer added successfully!")

def search_customer():
    search_name_value = search_name.get()
    for customer in list_customer:
        if search_name_value in customer.name:
            showinfo("Search result", 
                     f"Customer name: {customer.name} - Email: {customer.email} - Number: {customer.number} - Check in date: {customer.check_in} - Check out date: {customer.check_out} - Total: {customer.total}")
            break
                                
def update_customer():            
    # lấy thông tin từ các entry
    check_in_value = checkin_date
    check_out_value = checkout_date
    total_value = total
    name_value = name.get()
    email_value = email.get()
    number_value = number.get()
    
    for customer in list_customer:
        if name_value == customer.name:
            customer.email = email_value
            customer.number = number_value
            showinfo("Update result", f"Update information for customer {customer.name} successfully!")
            
def delete_customer():
    search_name_value = search_name.get()
    for i in range(len(list_customer)):
        if search_name_value == list_customer[i].name:
            del list_customer[i]
            showinfo("Action result", "Delete customer successfully")
            break    
            
        
# Add Button and Label
Button(window, text="Check in", command=check_in).grid(row=10, column=1)
Button(window, text="Check out", command=check_out).grid(row=10, column=2)
Button(window, text="Calculate days", command=calculate_days).grid(row=10, column=3)

checkin_label = Label(window, text="Check in")
checkin_label.grid(row=11, column=3)

checkout_label = Label(window, text="Check out")
checkout_label.grid(row=12, column=3)

total_label = Label(window, text="")
total_label.grid(row=13, column=3)
    
#Button
submit_btn = Button(window, text='Submit', command=add_customer)
search_btn = Button(window, text='Search', command=search_customer)
delete_btn = Button(window, text='Delete', command=delete_customer)
update_btn = Button(window, text="Update", command=update_customer)

# Display:
intro_lbl.grid(row=0, column=0, columnspan=2)
my_name_lbl.grid(row=1, column=0)
name.grid(row=1, column=1)
email_lbl.grid(row=2, column=0)
email.grid(row=2, column=1)
number_lbl.grid(row=3, column=0)
number.grid(row=3, column=1)


submit_btn.grid(row=4, column=1)

#Tracking customer
search_name_lbl.grid(row=1, column=3)
search_name.grid(row=2, column=3)
search_btn.grid(row=2, column=4)
update_btn.grid(row=3, column=4)
delete_btn.grid(row=4, column=4)


window.mainloop()
```

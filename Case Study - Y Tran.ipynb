```python
class Customer:
    def __init__(self, name, email, number, check_in, check_out):
        self.name = name
        self.email = email
        self.number = number
        self.check_in = check_in
        self.check_out = check_out
        
    def calc_total(self):
        self.total = (self.check_out - self.check_in) * 10000
```

```python
from tkinter import *
from tkinter.messagebox import showinfo

window = Tk()
window.title('STORE')
window.geometry('800x300')

# Label
intro_lbl = Label(window, text='Storge')
check_out_title_lbl = Label(window, text='Check out')

my_name_lbl = Label(window, text='Name')
email_lbl = Label(window, text='Email')
number_lbl = Label(window, text ='Phone number') 
check_in_lbl = Label(window, text = 'Check in time (hour)')
check_out_lbl = Label(window, text = 'Check out time (hour)')
search_name_lbl = Label(window, text = 'Search customer')

#Entry
name = Entry(window)
email = Entry(window)
number = Entry(window)
check_in = Entry(window)
check_out = Entry(window)

search_name = Entry(window)

#List_patient được sử dụng để lưu trữ thông tin các bệnh nhân
list_customer = []

def add_customer():
    #Lấy thông tin từ các entry:
    check_in_value = float(check_in.get())
    check_out_value = float(check_out.get())
    name_value = name.get()
    email_value = email.get()
    number_value = number.get()
    
    #Khởi tạo đối tượng:
    customer = Customer(name_value, email_value, number_value, check_in_value, check_out_value)
    
    #Thêm đối tượng vào danh sách:
    list_customer.append(customer)
    
    showinfo("Add Customer", "Customer added successfully!")

def search_customer():
    search_name_value = search_name.get()
    for customer in list_customer:
        if search_name_value in customer.name:
            showinfo("Search result", 
                     f"Customer name: {customer.name} - Email: {customer.email} - Number: {customer.number} - Check in: {customer.check_in} - Check out: {customer.check_out}")
            break
                                
def update_customer():            
    # lấy thông tin từ các entry
    check_in_value = float(check_in.get())
    check_out_value = float(check_out.get())
    name_value = name.get()
    email_value = email.get()
    number_value = number.get()
    
    for customer in list_customer:
        if name_value == customer.name:
            customer.email = email_value
            customer.number = number_value
            customer.check_in = check_in_value
            customer.check_out = check_out_value
            customer.calc_total()
            showinfo("Update result", f"Update information for customer {customer.name} successfully!")
            
def delete_customer():
    search_name_value = search_name.get()
    for i in range(len(list_customer)):
        if search_name_value == list_customer[i].name:
            del list_customer[i]
            showinfo("Action result", "Delete customer successfully")
            break    
#Button
submit_btn = Button(window, text='Submit', command=add_customer)
search_btn = Button(window, text='Search', command=search_customer)
delete_btn = Button(window, text='Delete', command=delete_customer)
update_btn = Button(window, text="Update", command=update_customer)

# Display:
intro_lbl.grid(row=0, column=0, columnspan=2)
my_name_lbl.grid(row=1, column=0)
name.grid(row=1, column=1)
email_lbl.grid(row=2, column=0)
email.grid(row=2, column=1)
number_lbl.grid(row=3, column=0)
number.grid(row=3, column=1)
check_in_lbl.grid(row=4, column=0)
check_in.grid(row=4, column=1)
check_out_lbl.grid(row=5, column=0)
check_out.grid(row=5, column=1)

submit_btn.grid(row=6, column=1)

#Tracking customer
search_name_lbl.grid(row=8, column=0)
search_name.grid(row=8, column=1)
search_btn.grid(row=8, column=2)
update_btn.grid(row=9, column=2)
delete_btn.grid(row=10, column=2)


window.mainloop()
```
